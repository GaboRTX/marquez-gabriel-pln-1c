## Ejercicio Práctico 1: Analizador Simple de Frases

> **Consigna:** Escribí un programa que reciba una frase del usuario y evalúe dos condiciones:
> 1. Si la frase contiene la palabra `"Python"`.
> 2. Si la frase es corta (menos de 20 caracteres).
>
> El programa debe imprimir un mensaje diferente para cada situación.

**Pasos sugeridos:**

1.  Pedí al usuario una frase usando `input()`.
2.  Medí su longitud con `len()`.
3.  Usá `if`, `elif` y `else` para manejar cada caso.
4.  Usá el operador `in` para verificar si una palabra está dentro del texto.


---

```python
# 1 Pedir al usuario que ingrese una frase
frase_usuario = input("Ingresa una frase: ")

# 2 Medir la longitud de la frase
longitud_frase = len(frase_usuario)
print(f"La frase tiene {longitud_frase} caracteres.")

# 3 Contiene la palabra "Python" SI/NO
# .lower() para que encuentre "python" aunque el usuario lo escriba en minusculas
if "python" in frase_usuario.lower():
    print("Tu frase habla sobre Python.")
else:
    print("Tu frase no menciona a Python.")

# 4 Evaluar si la frase es corta (menos de 20 caracteres)
if longitud_frase < 20:
    print("Esta es una frase corta.")
else:
    print("Esta es una frase larga.")
```

- *"¿Por qué `'python' in frase` y `'Python' in frase` pueden dar distintos resultados? ¿Cómo se soluciona?"*
>Por el case sensitivity, se soluciona normalizando (pasando el resultado del imput a minuscula).
- *"¿Puedo combinar la condición de longitud y la de pertenencia en un solo `if`? ¿Cómo?"*
>Si, con los operadors logicos and y or.

## Ejercicio Práctico 2: Contador de Frecuencias

> **Consigna:** A partir del texto dado, escribí un programa que lo convierta a minúsculas, lo divida en palabras y cuente cuántas veces aparece cada una.
>
> El resultado debe ser un diccionario donde las claves sean las palabras y los valores sean sus frecuencias.

**Pasos sugeridos:**

1.  Convertí el texto a minúsculas con `.lower()`.
2.  Dividilo en palabras con `.split()`.
3.  Recorrí la lista con un `for`.
4.  Para cada palabra, actualizá su contador en el diccionario.

```python
# 1 Pedimos el ingreso por teclado
texto_usuario = input("Ingresar texto para analizar: ")

# 2 Convertimos a minúsculas
texto_limpio = texto_usuario.lower()

# 3 Dividimos en palabras
palabras = texto_limpio.split()

# 4 Contador con diccionario
frecuencias = {}

for p in palabras:
    # Si la palabra ya está en el diccionario, sumamos 1
    if p in frecuencias:
        frecuencias[p] += 1
    # Si no está, la inicializamos en 1
    else:
        frecuencias[p] = 1

# 5 Resultado ordenado y prolijo
print("\n--- Resultado del Análisis ---")
for palabra, cantidad in frecuencias.items():
    print(f"La palabra '{palabra}' aparece {cantidad} veces.")
```
