## 3. ProgramaciĂłn Orientada a Objetos (POO): Un CĂłdigo mĂĄs Inteligente

La POO es un paradigma para organizar el cĂłdigo. La idea es simple: en lugar de tener datos y funciones por separado, los agrupamos en "objetos" que tienen tanto los datos (atributos) como las funciones que operan sobre esos datos (mĂŠtodos).

* **Clase**: Es el "plano" o la "receta". Define cĂłmo serĂĄn los objetos.
* **Objeto (o instancia)**: Es la "casa" construida a partir del plano. Es una entidad real con la que interactuamos.
* **`__init__(self, ...)`**: El "constructor". Es un mĂŠtodo especial que se ejecuta al crear un nuevo objeto para inicializar sus atributos.
* **`self`**: Se refiere al objeto mismo. AsĂ­ es como el objeto accede a sus propios datos y mĂŠtodos.

Vamos a crear una clase `ProcesadorTexto` que encapsule un texto y todas las operaciones que queremos hacer con ĂŠl.

```python
# DefiniciĂłn ĂNICA de la clase. Esta celda es la referencia para todo el notebook.
class ProcesadorTexto:
    """Encapsula un texto y expone operaciones de anĂĄlisis lingĂźĂ­stico bĂĄsico."""

    def __init__(self, texto):
        # Nota sobre limpieza: strip() solo elimina caracteres en los extremos del string.
        # Para limpiar puntuaciĂłn en todo el texto usamos replace(), igual que en limpiar_texto().
        print(f"Creando procesador para: '{texto[:40]}...'")
        self.texto_original = texto
        caracteres_a_eliminar = ['.', ',', '!', '?', ';', ':', 'Âż', 'ÂĄ', '"', "'", '(', ')']
        texto_limpio = texto.lower()
        for char in caracteres_a_eliminar:
            texto_limpio = texto_limpio.replace(char, '')
        self.texto_limpio = texto_limpio
        self.palabras = self.texto_limpio.split()

    def contar_palabras(self):
        """Devuelve el total de palabras (con repeticiones)."""
        return len(self.palabras)

    def contar_palabras_unicas(self):
        """Devuelve la cantidad de palabras distintas. `set` elimina duplicados."""
        return len(set(self.palabras))

    def calcular_frecuencia(self):
        """Devuelve un diccionario con la frecuencia de cada palabra."""
        frecuencia = {}
        for palabra in self.palabras:
            frecuencia[palabra] = frecuencia.get(palabra, 0) + 1
        return frecuencia
```

```python
# --- Uso de la clase ---
# Creamos un objeto (una instancia) de nuestra clase
procesador = ProcesadorTexto("Python para NLP es genial porque Python es versĂĄtil.")

# Ahora podemos usar sus mĂŠtodos
print(f"Total de palabras: {procesador.contar_palabras()}")
print(f"Palabras Ăşnicas: {procesador.contar_palabras_unicas()}")
print(f"Frecuencia de palabras: {procesador.calcular_frecuencia()}")
```

La clase `ProcesadorTexto` ya estĂĄ definida con tres mĂŠtodos. Tu tarea es agregarle uno nuevo.

> **Consigna:** AĂąadĂ­ el mĂŠtodo `calcular_longitud_media(self)` a la clase. Este mĂŠtodo debe calcular y devolver la longitud promedio de las palabras en el texto.
>
> **Pista:** la longitud promedio = suma de las longitudes de todas las palabras / cantidad total de palabras.

**Pasos sugeridos:**

1.  CopiĂĄ la clase completa desde la secciĂłn 3 en la celda de abajo.
2.  Agregale el mĂŠtodo `calcular_longitud_media(self)`.

```python
class ProcesadorTexto:
    """Encapsula un texto y expone operaciones de anĂĄlisis lingĂźĂ­stico basico."""

    def __init__(self, texto):
        # Nota sobre limpieza: strip() solo elimina caracteres en los extremos del string.
        # Para limpiar puntuaciĂłn en todo el texto usamos replace(), igual que en limpiar_texto().
        print(f"Creando procesador para: '{texto[:40]}...'")
        self.texto_original = texto
        caracteres_a_eliminar = ['.', ',', '!', '?', ';', ':', 'Âż', 'ÂĄ', '"', "'", '(', ')']
        texto_limpio = texto.lower()
        for char in caracteres_a_eliminar:
            texto_limpio = texto_limpio.replace(char, '')
        self.texto_limpio = texto_limpio
        self.palabras = self.texto_limpio.split()

    def contar_palabras(self):
        """Devuelve el total de palabras (con repeticiones)."""
        return len(self.palabras)

    def contar_palabras_unicas(self):
        """Devuelve la cantidad de palabras distintas. `set` elimina duplicados."""
        return len(set(self.palabras))

    def calcular_frecuencia(self):
        """Devuelve un diccionario con la frecuencia de cada palabra."""
        frecuencia = {}
        for palabra in self.palabras:
            frecuencia[palabra] = frecuencia.get(palabra, 0) + 1
        return frecuencia

    # --- TU CĂDIGO ACĂ ---
    def calcular_longitud_media(self):
        """Calcula la longitud promedio de las palabras en el texto."""
        if not self.palabras: # ValidaciĂłn por si el texto estĂĄ vacĂ­o
            return 0
        
        suma_longitudes = 0
        for palabra in self.palabras:
            suma_longitudes += len(palabra)
        
        promedio = suma_longitudes / len(self.palabras)
        return promedio

# --- ProbĂĄ tu mĂŠtodo ---
procesador_prueba = ProcesadorTexto("Los modelos de lenguaje son complejos.")
long_media = procesador_prueba.calcular_longitud_media()

# Usamos :.2f para mostrar solo 2 decimales
print(f"La longitud media de las palabras es: {long_media:.2f}")
```
